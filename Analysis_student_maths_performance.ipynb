```python
#The distribution of family size
df_student_math['family_size'].value_counts()
```

```python
#Use spearman find the correlation between family and students performance in the final_grade
#Use spearman because i'm looking for correlation between an int and an object value
#the result shows a weak positive relationship between family_size and final_grade
from scipy.stats import spearmanr
spearmanr(df_student_math['family_size'],df_student_math['final_grade'])
```

```python
#find the distribution of free time
df_student_math['free_time'].value_counts()
```

```python
#find the correlation between free_time and final_grade
##the result shows a weak positive relationship between free_time and final_grade
df_student_math['free_time'].corr(df_student_math['final_grade'])
```

```python
#find the distribution of address_type
df_student_math['address_type'].value_counts
```

```python
#the corelation between address_type and final_grade
#the result shows a weak positive relationship between address_type and final_grade
spearmanr(df_student_math['address_type'],df_student_math['final_grade'])
```
